```julia
using Plots
using CSV
using DataFrames
```

```julia
data = CSV.read("output/1.txt", DataFrame, header=false)
```

```julia
nn = length(data.Column1)
nx = Int(sqrt(nn))
```

```julia
u = zeros(nx, nx)
for j in 1:nx
    for i in 1:nx
        u[i, j] = data.Column5[(j-1)*nx + i]
    end
end
```

```julia
heatmap(u)
```
